# import pandas

```python
import pandas as pd
```

# read csv file from pandas

```python
df = pd.read_csv("store.csv")
```

#Preview Data

```python
df.head
```

#Clean Data
### rename colums

```python
col_name = list(df.columns)
```

## list comprehension

```python
clean_col = [name.lower().replace(" ","_").replace("-","_") for name in col_name]
```

```python
df.columns = clean_col
```

Preview clean data

```python
df.head()
```

# Data tranfomation

```python
df[df["country"]== "United States"][["customer_name","country"]].head()
```

# Query

```python
df.query("city == 'Los Angeles' and category == 'Furniture'").head()
```

# Aggregate data

```python
df.groupby(["segment","region"])[["sales","profit"]].agg(["sum","mean","std"]).reset_index().head()
```

# write csv

```python
df.to_csv("new_store.csv")
```
